```python
import os
import pandas as pd

i=0

def plotter(beta):

    random_color=list(np.random.choice(range(255),size=3))

    # Define the directory path
    
    betaAsString = str(beta)
    
    # directory = 'task5_data/0.2/'
    directory = 'task5_data/' + betaAsString + '/'

    # Initialize an empty list to store DataFrames
    dataframes = []

    # Loop through all files in the directory
    for filename in os.listdir(directory):
        # Check if the file is a regular file
        if os.path.isfile(os.path.join(directory, filename)):
            # Open the file with Pandas
            df = pd.read_csv(os.path.join(directory, filename))  # Assuming the file is CSV, change the function if needed
            
            r_correlation = df['r_correlation'][0]

            G = np.mean(df['G'])
            
            # G_err = 

            # G = df['G']
            
            
            
            color=[x/255 for x in random_color]
            
            plt.plot(r_correlation, G, linestyle="", marker='o', color=color)
            
            # Append the DataFrame to the list
            dataframes.append(df)

    # Concatenate all DataFrames into one
    combined_df = pd.concat(dataframes, ignore_index=True)

    # Now combined_df contains all data from all files in the directory


    plt.plot(r_correlation, G, linestyle="", marker='o', label=betaAsString, color=color)
    
    plt.legend()

    plt.xlabel("Distance between particles $r$")
    plt.ylabel("Correlation function $G_{ij}$")

    plt.show()
    
    return
```

```python
plotter(0.2)
```

```python
directory= 'task5_data'

# Loop through all items in the directory
for item in os.listdir(directory):
    # Check if the item is a directory
    if os.path.isdir(os.path.join(directory, item)):
        plotter(item)
```
